```python
import pandas as pd
import numpy as np
import random as rd
import seaborn as sn
import matplotlib.pyplot as plt
# from ucimlrepo import fetch_ucirepo 
```

```python
df = pd.read_csv("Data/default of credit card clients.csv")
```

```python
# Look at first 5 rows
df.head()
```

```python
# drop first row
defaults = df.drop(index=0)
defaults = defaults.drop("Unnamed: 0", axis = 1)
defaults.head()
```

```python
#Convert datatypes from string to ints
defaults = defaults.astype(int)
type(defaults.iloc[1,1])
```

```python
# Look at distributions of data
defaults.hist(bins = 20, figsize=(20,15))
plt.show()
```

```python
defaults_transformed = defaults
fields_to_transform = ["X1", "X5", "X15","X3", "X22"]
defaults_transformed[fields_to_transform] = defaults_transformed[fields_to_transform].apply(lambda x: np.log(x))
defaults_transformed.hist(bins = 20, figsize = (20,15))
```
